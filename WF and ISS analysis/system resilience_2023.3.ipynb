# Climate data analysis

```python
province_list = ["Heilongjiang", "Jilin", "Liaoning", "Jiangsu", "Anhui", "Hubei", "Sichuan", "Guizhou", "Yunnan",
                 "Jiangxi2S", "Hunan2S", "Hunan1S", "Jiangxi1S", "Zhejiang", "Fujian1S", "Guangdong", "Guangxi", "Hainan", "Fujian2S"]
base_path = "C:/Users/Dell/PycharmProjects/WQQM_PIDU/wf_new1/"
# 读取降雨数据
prcp_dic = {}
for province in province_list:
    prcp_file = "{}{}/totally decentralized/prcp_A.txt".format(base_path, province)
    prcp_dic[province] = pd.read_csv(prcp_file, sep="\t", header=0, skiprows=1, index_col=0)
prcp_A = pd.concat(list(prcp_dic.values()), axis=1)
prcp_A.transpose().describe()
```

```python
# 读取PET数据
pet_dic = {}
for province in province_list:
    pet_file = "{}{}/totally decentralized/PET_d.txt".format(base_path, province)
    et_file = "{}{}/totally decentralized/ET_d.txt".format(base_path, province)
    pet_d = pd.read_csv(pet_file, sep="\t", header=0, index_col=0, parse_dates=True)
    et_d = pd.read_csv(et_file, sep="\t", header=0, skiprows=1, index_col=0, parse_dates=True)
    pet_filter = pet_d.loc[et_d.dropna().index, :]
    pet_dic[province] = pet_filter.resample("A", kind="period").sum()
pet_A = pd.concat(list(pet_dic.values()), axis=1)
pet_A.transpose().describe()
```

```python
pet_A1 = pet_A.reset_index()
pet_A1["date1"] = pet_A1["date"].astype("str").astype("int")
pet_A1.drop("date", axis=1, inplace=True)
pet_A1.rename(columns= {"date1": "date"}, inplace=True)
pet_A1.set_index("date", inplace=True)
pet_A1.index
```

```python
AI_df = pet_A1 / prcp_A
AI_df.transpose().describe()
```

# Calculate system irrigation self-sufficiency for the irrigation event scale

```python
input_path = "C:/Users/Dell/PycharmProjects/WQQM_PIDU/input_data/"
ponding_day_dic = {"Heilongjiang": ["5-5"], "Jilin": ["5-5"], "Jiangsu": ["5-27"], "Anhui": ["5-27"], 
                   "Hunan1S": ["5-27"], "Hunan2S": ["4-16", "7-28"], "Sichuan": ["5-27"], "Guizhou": ["5-27"], "Yunnan": ["5-27"],
                   "Fujian1S": ["5-27"], "Fujian2S": ["4-10", "7-25"], "Guangdong": ["4-10", "7-25"], "Guangxi": ["4-10", "7-25"], "Hainan": ["4-10", "7-25"], 
                   "Hubei": ["5-27"], "Jiangxi1S": ["5-27"], "Jiangxi2S": ["4-16", "7-28"], "Liaoning": ["5-10"], "Zhejiang": ["6-15"]}
ISS_NRD_dic = {}
for province in province_list:
    # 读取灌溉事件尺度（日）的灌溉数据，计算灌溉自给率
    Iall_file = "{}{}/totally decentralized/I_d.txt".format(base_path, province)
    Ipond_file = "{}{}/totally decentralized/I_pond_d.txt".format(base_path, province)
    Iall_d = pd.read_csv(Iall_file, sep="\t", header=0, skiprows=1, index_col=0, parse_dates=True)
    Ipond_d = pd.read_csv(Ipond_file, sep="\t", header=0, skiprows=1, index_col=0, parse_dates=True)
    Iall_d[(Iall_d.shift(1) > 0) & (Iall_d < 10.0)] = np.NAN
    Iall_d[Iall_d < 5.0] = np.NAN
    Irrg_SS = Ipond_d / Iall_d * 100.0
    ponding_dates = []
    for year in range(1988, 2018, 1):
        for date in ponding_day_dic[province]:
            ponding_dates.append(str(year) + "-" + date)
    Irrg_SS.loc[ponding_dates, :] = np.NAN
    # 计算前期no rain days
    if ("1S" in province) | ("2S" in province):
        province1 = province[:-2]
    else:
        province1 = province
    prcp_d_file = "{}prcp_{}.txt".format(input_path, province1)
    prcp_d = pd.read_csv(prcp_d_file, sep="\t", header=0, index_col=0, parse_dates=True)
    NRD = pd.DataFrame(index=prcp_d.index, columns=prcp_d.columns, data=np.zeros((prcp_d.shape)).astype(np.int8))
    NRD[prcp_d.shift(1) < 1] = 1
    i = 1
    num_i = np.count_nonzero(NRD == i)
    while num_i > 0:
        NRD[NRD == i] = NRD + NRD.shift(i)
        i += 1
        num_i = np.count_nonzero(NRD == i)
    print(province, i)
    # 匹配no rain days和ISS
    ISS_arr = Irrg_SS.values.flatten()
    ISS_arr = ISS_arr[np.logical_not(np.isnan(ISS_arr))]
    NRD_arr = NRD[Irrg_SS >= 0].values.flatten()
    NRD_arr = NRD_arr[np.logical_not(np.isnan(NRD_arr))]
    ISS_NRD_df = pd.DataFrame({"ISS": ISS_arr, "NRD": NRD_arr, "Local storage volume (mm)": province_dp_volume_TD[province1], "province": province1})
    ISS_NRD_dic[province] = ISS_NRD_df
ISS_NRD_df = pd.concat(list(ISS_NRD_dic.values()), axis=0, ignore_index=True)
ISS_NRD_df.describe()
```
